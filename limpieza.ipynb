```python
import pandas as pd

# Archivo de estaciones actualizado con 'id_estacion', 'latitud' y 'longitud'
estaciones_file = pd.read_excel(r"C:\Users\nicke\Labo1\estaciones_actualizado.xlsx")

# Archivo original de precios con 'latitud' y 'longitud' añadidos
precios = pd.read_excel(r"C:\Users\nicke\Labo1\precios.xlsx")

# Asegurarse de que el id_estacion sea entero
estaciones['id_estacion'] = estaciones['id_estacion'].astype(int)

# Realizar un merge para añadir el 'id_estacion' al archivo de precios
precios_merged = precios.merge(
    estaciones[['id_estacion', 'Latitud', 'Longitud']],
    on=['Latitud', 'Longitud'],
    how='left'
)

# Diccionario de combustibles y sus IDs
combustibles = {
    "Precio gasolina 95 E5": 1,
    "Precio gasolina 95 E10": 2,
    "Precio gasolina 95 E5 Premium": 3,
    "Precio gasolina 98 E5": 4,
    "Precio gasolina 98 E10": 5,
    "Precio gasóleo A": 6,
    "Precio gasóleo Premium": 7,
    "Precio gasóleo B": 8,
    "Precio gasóleo C": 9,
    "Precio bioetanol": 10,
    "Precio biodiésel": 11,
    "Precio gases licuados del petróleo": 12,
    "Precio gas natural comprimido": 13,
    "Precio gas natural licuado": 14,
    "Precio hidrógeno": 15,
    "Precio gasóleo de uso marítimo": 16
}

# Normalizar los datos de precios
precios_normalizados = []
for combustible, id_combustible in combustibles.items():
    temp = precios_merged[['id_estacion', 'Toma de datos', combustible]].copy()
    temp.rename(columns={combustible: 'precio', 'Toma de datos': 'fecha_toma_datos'}, inplace=True)
    temp['id_combustible'] = id_combustible
    precios_normalizados.append(temp)

# Concatenar todos los precios en un solo DataFrame
precios_normalizados = pd.concat(precios_normalizados)

# Reemplazar valores NaN en 'id_estacion' por 0
precios_normalizados['id_estacion'].fillna(0, inplace=True)

# Asegurarse de que el id_estacion sea entero en el DataFrame final
precios_normalizados['id_estacion'] = precios_normalizados['id_estacion'].astype(int)

# Reemplazar valores nulos en otros campos por 0
precios_normalizados.fillna(0, inplace=True)

# Guardar el resultado en un archivo Excel
precios_normalizados.to_excel(r"C:\Users\nicke\Labo1\precios_actualizado.xlsx", index=False)

print("Archivo generado: precios_actualizado.xlsx")
```

```python
# Archivo de estaciones actualizado con 'id_estacion', 'latitud' y 'longitud'
estaciones = pd.read_excel(r"C:\Users\nicke\Labo1\estaciones_actualizado.xlsx")

# Archivo de horarios con 'latitud' y 'longitud'
horarios = pd.read_excel(r"C:\Users\nicke\Labo1\horarios.xlsx")

# Asegurarse de que el id_estacion sea entero
estaciones['id_estacion'] = estaciones['id_estacion'].astype(int)

# Realizar un merge para añadir el 'id_estacion' al archivo de horarios
horarios_merged = horarios.merge(
    estaciones[['id_estacion', 'Latitud', 'Longitud']],
    on=['Latitud', 'Longitud'],
    how='left'
)

# Reemplazar valores NaN en 'id_estacion' por 0
horarios_merged['id_estacion'].fillna(0, inplace=True)

# Asegurarse de que el id_estacion sea entero
horarios_merged['id_estacion'] = horarios_merged['id_estacion'].astype(int)

horarios_merged= horarios_merged.drop(columns = ['Longitud','Latitud'])
# Guardar el resultado en un archivo Excel
output_file = r"C:\Users\nicke\Labo1\horarios_actualizado.xlsx"
horarios_merged.to_excel(output_file, index=False)

print(f"Archivo generado: {output_file}")
```
